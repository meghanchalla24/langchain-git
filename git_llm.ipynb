```python
import os
from dotenv import load_dotenv


from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
api_key = os.getenv("GOOGLE_API_KEY")
```

```python
analyser_model = ChatGoogleGenerativeAI(
    model="gemini-3-pro-preview", # Note: Updated to a valid model name
    temperature=0.5,  
    max_tokens=4096,
    vertexai=False,
    api_key=api_key 
)
```

```python
response = analyser_model.invoke("What is the capital of France?")
```

```python
print(response.text)
```

```python
import subprocess
```

```python
def github_commit_tool(commit_message):
    subprocess.run(["git","pull"])
    subprocess.run(["git", "add", "."])
    subprocess.run(["git", "commit", "-m", commit_message])
    subprocess.run(["git", "push", "origin", "main"])
    print("Commit pushed successfully")
```

## Langchain Tool

```python
from langchain.tools import tool

@tool
def github_commit_tool(commit_message: str):
    """Commit and push code to GitHub."""
    
    subprocess.run(["git","pull"])
    subprocess.run(["git","add","."])
    subprocess.run(["git","commit","-m",commit_message])
    subprocess.run(["git","push","origin","main"])

    return "Commit pushed successfully"
```

```python
tools = [github_commit_tool]
```

```python
from langchain.agents import create_agent
```

```python

agent = create_agent(
    model=analyser_model,
    tools=tools,
    # You can add a system prompt to guide the agent's behavior
    system_prompt="You are a helpful assistant that can use tools.",
    verbose=True
)
```

```python
user_prompt = "Make commit to the github repo with the message : 'commit from the llm'"
```

```python
# The agent executor is implied when using agent.invoke
response = agent.invoke({"messages": [{"role": "user", "content": user_prompt}]})


```
