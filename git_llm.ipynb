```python
import os
from dotenv import load_dotenv


from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
api_key = os.getenv("GOOGLE_API_KEY")
```

```python
analyser_model = ChatGoogleGenerativeAI(
    model="gemini-3-pro-preview", # Note: Updated to a valid model name
    temperature=0.5,  
    max_tokens=4096,
    vertexai=False,
    api_key=api_key 
)
```

```python
response = analyser_model.invoke("What is the capital of France?")
```

## Langchain Tool

```python
from langchain.tools import tool

@tool
def github_commit_tool(commit_message):
    """This function uses he subprocess package to make commits to a github repo pre-defined"""
    subprocess.run(["git","pull"])
    subprocess.run(["git", "add", "."])
    subprocess.run(["git", "commit", "-m", commit_message])
    subprocess.run(["git", "push", "origin", "main"])
    print("Commit pushed successfully")
    return "success"
```

```python
tools = [github_commit_tool]
```

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(
    analyser_model,
    tools,
    # You can add a system prompt to guide the agent's behavior
    system_prompt="Make commit to the github repo with the message : 'commit from the llm'",
)
```
